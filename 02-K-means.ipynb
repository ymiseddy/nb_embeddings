### Steps

1. Start by randomly geberating centroids
2. Assign each vector to the closest centroid
3. Move the centroid to the center of mass (means of x by means of y)
4. Repeat 2 and 3 until no vectors change centroids

### Notes
- Occasionally a few points will flip back and forth causing an infinite loop
  - **Solution:** Have a hard limit to the number of iterations
- Random centroids can sometimes be close together interlocking them to the wrong cluster
  - **Solution:** Run the clustering multiple times.  Use a loss-function such as the Root Mean Square Error (RMSE)
    identify the best clustering.  After mutliple runs, pick the one with the smallest RMSE.
